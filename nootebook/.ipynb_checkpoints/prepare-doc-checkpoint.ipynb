```python
import time
import pickle
import random
import subprocess
from tqdm import tqdm
from src.utils import (display_base64_image, 
                       load_config,
)
```

```python
general_config = load_config('config.yaml')
models_names = general_config['models']
```

```python
from unstructured.partition.pdf import partition_pdf

data_path = '/kaggle/input/data-pdf/k24-493.pdf'

chunks = partition_pdf(
    filename=data_path,
    infer_table_structure=True,            
    strategy="hi_res",             
    extract_image_block_types=["Image"],   
    extract_image_block_to_payload=True,   
    max_characters=10000,                  
    combine_text_under_n_chars=2000,       
    new_after_n_chars=6000,
)
```

```python
print(f'Length of extracted chunks: {len(chunks)}')
print(f'Unique types of extracted chunks: {set([str(type(el)) for el in chunks])}')
```

```python
tables, texts, images, titles, footer, header, narrative_texts, list_items = [], [], [], [], [], [], [], []

# Categorize each chunk based on its type
for chunk in tqdm(chunks, desc='Sorting chunks'):
    if "unstructured.documents.elements.Table" in str(type(chunk)):
        tables.append(str(chunk))
    elif "unstructured.documents.elements.NarrativeText" in str(type(chunk)):
        narrative_texts.append(str(chunk))
    elif "unstructured.documents.elements.Text" in str(type(chunk)):  # Text objects
        texts.append(str(chunk))
    elif "unstructured.documents.elements.Image" in str(type(chunk)):
        images.append(chunk)
    elif "unstructured.documents.elements.Title" in str(type(chunk)):
        titles.append(str(chunk))
    elif "unstructured.documents.elements.Footer" in str(type(chunk)):
        footer.append(str(chunk))
    elif "unstructured.documents.elements.Header" in str(type(chunk)):
        header.append(str(chunk))
    elif "unstructured.documents.elements.ListItem" in str(type(chunk)):
        list_items.append(str(chunk))
```

```python
base64_images = [el.metadata.image_base64 for el in images]

for image in base64_images[:4]:
    display_base64_image(image)
```

```python
base64_images = base64_images[4:]
display_base64_image(base64_images[0])
```

# Summarizing texts and tables

```python
from groq import Client
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
os.environ["GROQ_API_KEY"] = "your_api_key"
client = Client(api_key=os.getenv("GROQ_API_KEY"))
```

```python
TEXT_SUMMARY_MODEL_NAME = models_names['text_summary_model']
```

```python
# Prompt
prompt_text = """
You are an assistant tasked with summarizing tables and text.
Give a concise summary of the table or text.

Respond only with the summary, no additionnal comment.
Do not start your message by saying "Here is a summary" or anything like that.
Just give the summary as it is.

Table or text chunk: {element}
"""


prompt = ChatPromptTemplate.from_template(prompt_text)
model = ChatGroq(temperature=0.5, model=TEXT_SUMMARY_MODEL_NAME)
summarize_chain = {"element": lambda x: x} | prompt | model | StrOutputParser()
```

```python
batch_size = 29
text_summaries = []

for i in tqdm(range(0, len(narrative_texts), batch_size), desc='Summarizing texts'):
    batch = narrative_texts[i:i + batch_size]
    text_summaries += summarize_chain.batch(batch, {"max_concurrency": 3})
    time.sleep(60)

table_summaries = summarize_chain.batch(tables, {"max_concurrency": 3})
```

```python
random_num = random.randint(0, len(tables) - 1)

print(f'Original table:\n\t{tables[random_num]}')
print(f'Summary:\n\t{table_summaries[random_num]}')
```

```python
random_num = random.randint(0, len(narrative_texts) - 1)

print(f'Original table:\n\t{narrative_texts[random_num]}')
print(f'Summary:\n\t{text_summaries[random_num]}')
```

```python
from langchain_ollama import OllamaLLM 

IMAGES_SUMMARY_MODEL_NAME = models_names['image_summary_model']

subprocess.run(["ollama", "run", IMAGES_SUMMARY_MODEL_NAME], 
               stdout=subprocess.DEVNULL, 
               stderr=subprocess.DEVNULL)

process = subprocess.Popen("ollama serve", shell=True) 
```

```python
prompt = """Describe the image in detail. For context,
            the image is part of a research paper. 
            Be specific about graphs, such as scatter plots, line plots."""
```

```python
images_summaries = []
llm = OllamaLLM(model=IMAGES_SUMMARY_MODEL_NAME)

for image in tqdm(base64_images, desc='Summarizing images'):
    llm_with_images = llm.bind(images=[image])
    summary = llm_with_images.invoke(prompt)
    images_summaries.append(summary)
```

# Vector Store

```python
from langchain_chroma  import Chroma
from langchain_ollama import OllamaEmbeddings
from langchain.storage import LocalFileStore, InMemoryByteStore
from langchain.retrievers.multi_vector import MultiVectorRetriever
```

```python
EMBEDDING_MODEL_NAME = models_names['embedding_model']
COLLECTION_NAME = general_config['parameters']['collection_name']

embeddings = OllamaEmbeddings(model=EMBEDDING_MODEL_NAME)

vectorstore = Chroma(embedding_function=embeddings, 
                     persist_directory='/kaggle/working/chroma_db')

subprocess.run(["ollama", "run", EMBEDDING_MODEL_NAME], stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL)
```

```python
id_key = "doc_id"

docstore = InMemoryByteStore()

retriever = MultiVectorRetriever(
    id_key=id_key,
    byte_store=docstore,
    vectorstore=vectorstore,
)
```

```python
from src.utils import (generate_ids, create_documents)

#texts
narrative_ids = generate_ids(narrative_texts)
narrative_docs, narrative_summary_docs = create_documents(narrative_texts, narrative_ids, text_summaries)
retriever.vectorstore.add_documents(narrative_summary_docs)
retriever.docstore.mset(list(zip(narrative_ids, narrative_docs)))

# tables
table_ids = generate_ids(tables)
table_docs, table_summary_docs = create_documents(tables, table_ids, table_summaries)
retriever.vectorstore.add_documents(table_summary_docs)
retriever.docstore.mset(list(zip(narrative_ids, table_docs)))

# images
image_ids = generate_ids(base64_images)
image_docs, image_summary_docs = create_documents(base64_images, image_ids, images_summaries)
retriever.vectorstore.add_documents(image_summary_docs)
retriever.docstore.mset(list(zip(narrative_ids, image_docs)))
```
